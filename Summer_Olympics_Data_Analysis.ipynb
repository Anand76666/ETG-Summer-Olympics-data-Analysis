# Importing Dataset

```python
df = pd.read_csv("summer.csv")
df.head()
```

### 1. In how many cities Summer Olympics is held so far?

```python
df["City"].describe()
```

```python
df["City"].value_counts(ascending=False).head()
```

```python
plt.figure(figsize=(12, 9))
df["City"].value_counts().plot(kind="barh")
```
